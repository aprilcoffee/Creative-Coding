# Image Convert


```python
from PIL import Image
import os

# Define input image path and output directory
input_image_path = "image/AfterWalkerEvans.jpeg"  # Replace with your input image file
output_directory = "convert"  # Replace with your desired output directory
```

```python
# Load the input image
img = Image.open(input_image_path)
img = img.convert("RGB")  # Ensure the image is in RGB mode for consistent format conversion
display(img)
```

```python

# Extract filename without extension
base_filename = os.path.splitext(os.path.basename(input_image_path))[0]

# Ensure the output subdirectory exists
output_subdir = os.path.join(output_directory, base_filename)
os.makedirs(output_subdir, exist_ok=True)

print(base_filename)
```

```python
# Save as PNG
png_path = os.path.join(output_subdir, "img.png")
img.save(png_path, format="PNG")

# Save as JPEG
jpeg_path = os.path.join(output_subdir, "img.jpeg")
img.save(jpeg_path, format="JPEG")

# Save as BMP
bmp_path = os.path.join(output_subdir, "img.bmp")
img.save(bmp_path, format="BMP")

# Save as GIF (check if animated)
gif_path = os.path.join(output_subdir, "img.gif")
if img.format == "GIF" and getattr(img, "is_animated", False):
    img.save(gif_path, save_all=True)
else:
    img.save(gif_path, format="GIF")

# Save as RAW (binary data)
raw_path = os.path.join(output_subdir, "img.raw")
with open(raw_path, 'wb') as raw_file:
    raw_file.write(img.tobytes())


# Print output paths
print(f"Saved images in the following formats:\n"
      f"PNG: {png_path}\n"
      f"JPEG: {jpeg_path}\n"
      f"BMP: {bmp_path}\n"
      f"RAW: {raw_path}\n"
      f"GIF: {gif_path}")
```
